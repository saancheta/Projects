```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=[20,10])

# input variables (i=medium, t=lens)
question= int(input("Enter 1 for one lens system. Enter 2 for two lens system. "))
if question == 1:
    dl= float(input("Thickness of lens 1:"))
    nl= float(input("Refractive index of lens 1:"))
    no= float(input ("Refractive index of the medium 1:")) 
    R1= float(input("Radius of curvature of first surface:"))
    R2 = float(input('Radius of curvature of second surface:'))
    Po= float(input("Object height:"))
    ni= float(input ("Refractive index of medium 2:")) #nt2
    #alphao= np.radians(float(input("Angle of incidence of incoming ray:")))
    do = float(input('Object distance from Vertex 1 of lens:'))
    
    if nl == no and no == ni:
        plt.axhline(y=Po, color='r', linestyle='-')
        plt.axvline(x=1, color='b', linestyle="--")
        plt.axvline(x=1+dl, color='b', linestyle="--")
        plt.show()
    
    
    #for effective focal length 1
    f_inv = ((nl/no)-1)*((1/R1)-(1/R2)+((((nl/no)-1)*dl)/(nl*R1*R2)))
    f = f_inv**(-1)

    #for h1 and h2
    h1 = -f*(nl-1)*dl/(R2*nl)
    h2 = -f*(nl-1)*dl/(R1*nl)

    D1 = (nl - no)/R1
    D2 = (ni - nl)/R2

    ####For A1
    a11= 1 - (D2*dl)/nl
    a21 = dl/nl
    a12 = -D1 - D2 + (D2*D1*dl)/ nl
    a22 = 1 - (D1*dl)/nl
    alpha= np.arange(-0.005,0.01,0.005)

    ##for final image from 1
    di = ((do)*f) /((do) - f)
    
    for i in alpha: 
        ##LENS 1
        y1 = do*i+Po
        y2 = a21*no*i+a22*(y1)
        alpha_new = (a11*no*i+a12*(y1))/ni
        y3 = di*(alpha_new) + y2
        #
        y_axis = [Po,y1,y2,y3]
        x_axis = [0, do+h1, do+dl+h2, do+dl+h2+di]
        print((round(y3,2)))
        plt.plot(x_axis,y_axis,"ro-")
        plt.arrow(0, 0, 0, Po, head_length=0.25*Po, length_includes_head=True)
        plt.arrow(do+dl+h2+di, 0, 0, y3, head_length=0.25*abs(y3), length_includes_head=True)
        plt.axvline(x=0,color="k",linestyle="--",linewidth=0.5)
        plt.axhline(y=0,color="k",linestyle="--",linewidth=0.5)
        plt.axvline(x=do+h1,color="b",linestyle="--",linewidth=0.5)
        plt.axvline(x=do+dl+h2,color="b",linestyle="--",linewidth=0.5) 
    plt.show()

elif question == 2:
    
    dl= float(input("Thickness of lens 1:"))
    nl= float(input("Refractive index of lens 1:"))
    no= float(input ("Refractive index of the medium 1:")) #ni1

    R1= float(input("Radius of curvature of first surface:"))
    R2 = float(input('Radius of curvature of second surface:'))
    Po= float(input("Object height:"))
    ni= float(input ("Refractive index of medium 2:")) #nt2
#alphao= np.radians(float(input("Angle of incidence of incoming ray:")))
    do = float(input('Object distance from Vertex 1 of lens:'))

    dd = float(input('Distance between lens1 and lens 2:')) 

#for lens 2
    dll = float(input('Thickness of lens 2:'))
    nll = float(input ("Refractive index of lens 2:")) 
    R3 =float(input("Radius of curvature of first surface of lens 2:"))
    R4 = float(input('Radius of curvature of second surface of lens 2:'))
    nii = float(input ("Refractive index of medium 3:")) #nt2

    if nl == no and no == ni and ni == nll and nll == nii:
        plt.axhline(y=Po, color='r', linestyle='-')
        plt.axvline(x=do+h1,color="b",linestyle="--",linewidth=0.5)
        plt.axvline(x=do+dl+h2,color="b",linestyle="--",linewidth=0.5) 
        plt.axvline(x=do+dl+h2+dd+h3,color="b",linestyle="--",linewidth=0.5)
        plt.axvline(x=do+dl+h2+dd+dll+h4,color="b",linestyle="--",linewidth=0.5) 
        plt.show()
    
#for effective focal length 1
    f_inv = ((nl/no)-1)*((1/R1)-(1/R2)+((((nl/no)-1)*dl)/(nl*R1*R2)))
    f = f_inv**(-1)

#for h1 and h2
    h1 = -f*(nl-1)*dl/(R2*nl)
    h2 = -f*(nl-1)*dl/(R1*nl)

    D1 = (nl - no)/R1
    D2 = (ni - nl)/R2

####For A1
    a11= 1 - (D2*dl)/nl
    a21 = dl/nl
    a12 = -D1 - D2 + (D2*D1*dl)/ nl
    a22 = 1 - (D1*dl)/nl
    alpha= np.arange(-0.005,0.01,0.005)

##for final image from 1
    di = ((do+h1)*f) /((do+h1) - f)

##LENS 2
# A2
    D3 = (nll - ni)/R3
    D4 = (nii - nll)/R4

    a112= 1 - (D4*dll)/nll
    a212 = dll/nll
    a122 = -D3 - D4 + (D4*D3*dll)/ nll
    a222 = 1 - (D3*dll)/nll

    alpha2=np.arange((np.pi/4)-0.01,0.80,0.01)

#for effective focal length 2
    ff_inv = (nll/ni-1)*((1/R3)-(1/R4)+((nll/ni-1)*dl/(nll*R3*R4)))
    ff = ff_inv**(-1)

#for h3 and h4
    h3 = -ff*(nll-1)*dll/(R4*nll)  
    h4 = -ff*(nll-1)*dll/(R3*nll)

#final final image
    do2 = dd-di-h2+h3
    dii = ((do2)*ff)/ ((do2)-ff)

    for i in alpha: 
    ##LENS 1
        y1 = -do*i+Po
        y2 = a21*no*i+a22*(y1)
        alpha_new = (a11*no*i+a12*(y1))
        y3 = di*(alpha_new) + y2
    #
        y_axis = [Po,y1,y2,y3]
        x_axis = [0, do+h1, do+dl+h2, do+dl+h2+di]
        #print((round(y3,2)))
        plt.plot(x_axis,y_axis,"ro-")
        plt.arrow(0, 0, 0, Po, head_length=0.25*Po, length_includes_head=True)
        plt.arrow(do+dl+h2+di, 0, 0, y3, head_length=0.25*abs(y3), length_includes_head=True)
        plt.axvline(x=0,color="k",linestyle="--",linewidth=0.5)
        plt.axhline(y=0,color="k",linestyle="--",linewidth=0.5)
        plt.axvline(x=do+h1,color="b",linestyle="--",linewidth=0.5)
        plt.axvline(x=do+dl+h2,color="b",linestyle="--",linewidth=0.5) 
    ##LENS 2
    for j in alpha2:
        y12 = -di*j+y3
        y22 = a212*ni*j+a222*(y12)
        alpha2_new = (a112*ni*j+a122*(y12))
        y32 = dii*(alpha2_new) + y22
    #
        y2_axis = [y3,y12,y22,y32]
        x2_axis = [do+dl+h2+di, do+dl+h2+dd+h3, do+dl+h2+dd+dll+h4, do+dl+h2+dd+dll+h4+dii]
        plt.plot(x2_axis,y2_axis,"go-")
    #plt.arrow(0, 0, 0, Po, head_length=0.25*Po, length_includes_head=True)
        plt.arrow(do+dl+h2+dd+dll+h4+dii, 0, 0, y32, head_length=0.25*abs(y32), length_includes_head=True)
        
        plt.axvline(x=do+dl+h2+dd+h3,color="b",linestyle="--",linewidth=0.5)
        plt.axvline(x=do+dl+h2+dd+dll+h4,color="b",linestyle="--",linewidth=0.5) 
    
    plt.show()

```
